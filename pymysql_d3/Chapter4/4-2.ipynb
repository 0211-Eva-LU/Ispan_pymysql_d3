```python
import pymysql
from configparser import ConfigParser

# 讀取 .env 檔案取得資料庫連線資訊
config = ConfigParser()
config.read('../Chapter1/config.ini')

# 建立資料庫連線
connection = pymysql.connect(
    host=config.get('DB', 'host'),
    user=config.get('DB', 'user'),
    password=config.get('DB', 'password'),
    port=config.getint('DB', 'port'),
    cursorclass=pymysql.cursors.DictCursor,
    database= 'superstore'
)

# 查看資料庫中的所有資料表
with connection.cursor() as cursor:
    cursor.execute("SHOW TABLES FROM superstore;")
    tables = cursor.fetchall()

print(tables)
```

```python
def sql_query(query):
    with connection.cursor() as cursor:
        cursor.execute(query)
        result = cursor.fetchall()
        
    return result
```

```python
def get_category_chart_data() -> list[dict[str, int]]:
    # 計算每個主類別的產品訂單數量
    category_query = """
    select  category as label ,count(category) as value
    from Products
    group by category;
    """
    data= sql_query(category_query)


    """
    將資料整理成
    [
        {'label': 'Category 1', 'value': 100},
        {'label': 'Category 2', 'value': 200},
        ...
    ]
    """


    return data
get_category_chart_data()
```

```python
def get_sub_category_chart_data(category):
    # 計算某個主類別，其子類別的產品訂單數量
    sub_category_query ="""
        select  sub_category as lable ,count(sub_category) as value
        from Products
        group by category,sub_category;"""
    data = sql_query(sub_category_query)
    """
    將資料整理成
    [
        {'label': 'Category 1', 'value': 100},
        {'label': 'Category 2', 'value': 200},
        ...
    ]
    """
    return data
get_sub_category_chart_data('Furniture')
```
