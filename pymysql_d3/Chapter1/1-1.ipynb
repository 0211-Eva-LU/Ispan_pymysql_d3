使用 cursorclass 修改回傳資料格式

```python
# 資料庫連線
from pprint import pprint
connection = pymysql.connect(
    host='localhost',
    user='root', 
    password='P@ssw0rd',
    port=3306,
    database='my_titanic',
    # 使用字典型游標
    cursorclass= pymysql.cursors.DictCursor
)

# 建立游標
with connection.cursor() as cursor:
    # 使用游標執行 SQL 語句
    sql = "SELECT * FROM full_passengers;"
    cursor.execute(sql)
    # 取得所有資料
    databases = cursor.fetchall()

pprint(databases[0])
```

查詢資料庫

```python
from pprint import pprint
connection = pymysql.connect(
    host='localhost',
    user='root', 
    password='P@ssw0rd',
    port=3306,
    database='my_titanic',
    # 使用字典型游標
    cursorclass= pymysql.cursors.DictCursor
)

# 建立游標
with connection.cursor() as cursor:
    # 使用游標執行 SQL 語句
    sql = "SELECT * FROM full_passengers;"
    cursor.execute(sql)
    # 取得所有資料
    databases = cursor.fetchall()

# pprint(databases[0])

Leonard_list = []


for i in databases:
    if 'Leonard' in i['pname'].split(",")[1] and i['sex'] == 'male': #startswith() 檢查開頭
        Leonard_list.append(i)

print(len(Leonard_list))
```
